## Quant Research: Mean Reversion Hypothesis Testing
### Framework: Statistical Cointegration and Z-Score Analysis

This notebook provides a skeleton for conducting mean reversion research on synthetic market data. It focuses on the mathematical foundation of Z-Score calculation and stationarity testing.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Institutional Plotting Configuration
plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.figsize'] = [12, 6]
```

### 1. Synthetic Data Generation
Simulating a cointegrated pair or a mean-reverting series for hypothesis validation.

```python
np.random.seed(42)
mu, sigma = 0, 0.01 # Standard Gaussian noise
noise = np.random.normal(mu, sigma, 1000)
prices = np.cumsum(noise) + 100 # Random walk

# Generating a mean-reverting spread
spread = np.random.normal(0, 1, 1000)
df = pd.DataFrame({'price': prices, 'spread': spread})
df.head()
```

### 2. Z-Score Calculation
The core metric for identifying statistical outliers in mean-reverting strategies.

```python
def calculate_zscore(series: pd.Series, window: int) -> pd.Series:
    """
    Calculates the rolling Z-Score of a given series.
    Formula: (x - rolling_mean) / rolling_std
    """
    rolling_mean = series.rolling(window=window).mean()
    rolling_std = series.rolling(window=window).std()
    z_score = (series - rolling_mean) / rolling_std
    return z_score

df['z_score'] = calculate_zscore(df['spread'], window=20)
df['z_score'].plot(title='Rolling Z-Score (Institutional Baseline)')
plt.axhline(0, color='black', alpha=0.5)
plt.axhline(2, color='red', linestyle='--')
plt.axhline(-2, color='green', linestyle='--')
plt.show()
```
